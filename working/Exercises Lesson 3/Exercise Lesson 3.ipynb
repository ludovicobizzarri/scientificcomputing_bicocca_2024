```python
import numpy as np
import matplotlib.pyplot as plt
```

# Planetary position

```python
distance = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])
period = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])
names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]
scale = period*0.1 # to scale the log-plot
```

```python
fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(111)

for a,P,n,s in zip(distance,period,names,scale):    
    ax.scatter(a,P)
    ax.text(a,P+s,n)

ax.set_xlabel(r'Distance from the sun [AU]')
ax.set_ylabel(r'Orbital period [yr]')
ax.set_title('Planetary period vs distance from the Sun')
ax.set_xscale('log')
ax.set_yscale('log')
```

# Frequent word plotting

```python
f = open('shore_leave.txt','r')

all_words = []

for line in f:
    line_words = line.split(' ')
    
    for word in line_words:
        all_words.append(word)
    
f.close()

# remove dots, commas and \n and put all the words lowercase
for i in range(len(all_words)):
    #print(single_word)
    if('.'in all_words[i]):
        all_words[i] = all_words[i].split('.')[0]
    
    if(','in all_words[i]):
        all_words[i] = all_words[i].split(',')[0]
        
    if('\n'in all_words[i]):
        all_words[i] = all_words[i].split('\n')[0]

    if(all_words[i].istitle()):
        all_words[i] = all_words[i][0].lower() + all_words[i][1:]
```

```python
word_list = {}

for single_word in all_words:
    if(single_word in word_list):
        word_list[single_word] += 1
        
    else:
        word_list.update({single_word: 1})
```
